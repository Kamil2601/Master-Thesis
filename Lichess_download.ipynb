```python
import requests
import re
import os
import pathlib
import time
```

```python
os.mkdir("./data")
os.mkdir("./data/popular_studies")
```

```python
STUDIES = "https://lichess.org/study/"

destination = "./data/popular_studies"

studies = 0

studies_ids = []

for page in range(26,40):
    try:
        response = requests.get(f"https://lichess.org/study/all/popular?page={page}")
    except BaseException:
        print(f"Error on page {page}")
        break
    page_studies_links = re.findall(r'[/]study[/]\w{8}["]', response.text)

    # print(page_studies_links)

    studies_ids += list(map(lambda x: x[7:-1], page_studies_links))
```

```python
f = open("./data/popular_studies_ids")
studies_ids = list(map(lambda x: x[:-1], f.readlines()))
```

```python
# while len(studies_ids) > len(os.listdir(""./data/popular_studies/))

for i, studyId in enumerate(studies_ids):
    if not os.path.exists(f"./data/popular_studies/{studyId}"):
        try:
            response = requests.get(f"https://lichess.org/api/study/{studyId}.pgn")
        except BaseException:
            print(f"Error on index {i}")
            break

        sleep_time = 2

        while response.text[0] == "{":
            time.sleep(sleep_time)
            sleep_time += 2
            response = requests.get(f"https://lichess.org/api/study/{studyId}.pgn")


        study_file = open(f"./data/popular_studies/{studyId}", "w")
        study_file.write(response.text)
```

```python
elite_database_path = "./data/Lichess_Elite_Database"
player_regex = r'[[]((Black)|(White))[ ]\"\w+\"[]]\n'
test = r'[[]White[ ]\"\w+\"[]]\n'

get_nick = lambda x: x[8:-3]

elite_players = set()

for path in pathlib.Path(elite_database_path).iterdir():
    if path.is_file():
        current_file = open(path, "r")

        players_in_file = set(list(map(get_nick, re.findall(test, current_file.read()))))

        elite_players |= players_in_file
        
elite_players = list(elite_players)
```
