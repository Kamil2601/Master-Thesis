```python
import sys
sys.path.append("..")

import chess
import chess.pgn
import io
from tqdm import tqdm
import pandas as pd
import pickle
from utils import get_comments, get_nags, load_games_from_pgn_file

from common.utils import save_to_sql
```

```python
chessbase_file = "../../data/ChessBase/annotated_games.pgn"
pickle_file = "../../data/ChessBase/annotated_games.bin"
books_directory = "../../data/libgen_books/Categories/winner_marked_or_last_move/"
```

## Chessbase

```python
chessbase_annotated_games = load_games_from_pgn_file(chessbase_file)
```

```python
len(chessbase_annotated_games)
```

```python
chessbase_comments = []

for game in tqdm(chessbase_annotated_games):
    chessbase_comments += get_comments(game)
```

```python
comments_df = pd.DataFrame(chessbase_comments, columns=["position", "move", "comment", "sentiment"]).astype(
    {"position": str, "move": str, "comment": str, "sentiment": int})
comments_df
```

```python
save_to_sql(comments_df, "../../chess.db", "chessbase_moves_with_comments", if_exists="replace")
```

```python
chessbase_annotated_moves = []

for game in chessbase_annotated_games:
    chessbase_annotated_moves += get_nags(game)

len(chessbase_annotated_moves)
```

```python
len(chessbase_annotated_moves)
```

```python
moves_df = pd.DataFrame(chessbase_annotated_moves, columns=["position", "move", "sentiment"]).astype({"position": "str", "move": str, "sentiment": int})
moves_df
```

```python
save_to_sql(moves_df, "../../chess.db", "chessbase_moves_with_nags", if_exists="replace")
```
